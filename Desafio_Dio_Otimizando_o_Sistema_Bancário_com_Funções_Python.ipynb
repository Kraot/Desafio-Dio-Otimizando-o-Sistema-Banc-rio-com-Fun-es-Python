```python
usuarios = [
    {"nome": "Marcos Silva", "data_nascimento": "15-04-1985", "cpf": "12345678901", "endereco": "Rua A, 100 - Centro - São Paulo/SP"}
]


contas = [
    {"agencia": "0001", "numero_conta": 1, "usuario": usuarios[0], "saldo": 0, "extrato": "", "numero_saques": 0, "emprestimo": 1000, "emprestimo_maximo": 1000}
]


def criar_usuario(usuarios):
    cpf = input("Informe o CPF (somente números): ")
    usuario_existente = next((usuario for usuario in usuarios if usuario["cpf"] == cpf), None)

    if usuario_existente:
        print("Usuário com este CPF já cadastrado!")
        return

    nome = input("Informe o nome completo: ")
    data_nascimento = input("Informe a data de nascimento (dd-mm-aaaa): ")
    endereco = input("Informe o endereço (logradouro, nro - bairro - cidade/sigla estado): ")

    usuarios.append({"nome": nome, "data_nascimento": data_nascimento, "cpf": cpf, "endereco": endereco})
    print("Usuário cadastrado com sucesso!")


def criar_conta_corrente(agencia, numero_conta, usuarios, contas):
    cpf = input("Informe o CPF do usuário: ")
    usuario = next((usuario for usuario in usuarios if usuario["cpf"] == cpf), None)

    if usuario:
        contas.append({"agencia": agencia, "numero_conta": numero_conta, "usuario": usuario, "saldo": 0, "extrato": "", "numero_saques": 0, "emprestimo": 1000, "emprestimo_maximo": 1000})
        print("Conta corrente criada com sucesso!")
        return numero_conta + 1
    else:
        print("Usuário não encontrado. Operação falhou!")
        return numero_conta


def listar_contas(contas):
    for conta in contas:
        print(f"\nAgência: {conta['agencia']}, Conta: {conta['numero_conta']}, Titular: {conta['usuario']['nome']}")


def selecionar_conta(contas):
    cpf = input("Informe o CPF do titular da conta: ")
    conta = next((conta for conta in contas if conta["usuario"]["cpf"] == cpf), None)

    if conta:
        print(f"Conta de {conta['usuario']['nome']} selecionada!")
        return conta
    else:
        print("Conta não encontrada!")
        return None


def depositar(conta, valor, /):
    if valor > 0:
        conta["saldo"] += valor

        if conta["emprestimo"] < conta["emprestimo_maximo"]:
            valor_reposto = min(valor, conta["emprestimo_maximo"] - conta["emprestimo"])
            conta["emprestimo"] += valor_reposto
            conta["saldo"] -= valor_reposto
            conta["extrato"] += f"Depósito: R$ {valor:.2f} (R$ {valor_reposto:.2f} reposto ao empréstimo)\n"
        else:
            conta["extrato"] += f"Depósito: R$ {valor:.2f}\n"

        print(f"Depósito realizado com sucesso! Saldo atual: R$ {conta['saldo']:.2f}, Empréstimo disponível: R$ {conta['emprestimo']:.2f}")
    else:
        print("Operação falhou! O valor informado é inválido.")


def sacar(*, conta, valor, limite, limite_saques):
    saldo_total = conta["saldo"] + conta["emprestimo"]

    if valor > saldo_total:
        print("Operação falhou! Você não tem saldo suficiente e o valor do empréstimo disponível não cobre o saque.")
    elif valor > limite:
        print("Operação falhou! O valor do saque excede o limite.")
    elif conta["numero_saques"] >= limite_saques:
        print("Operação falhou! Número máximo de saques excedido.")
    elif valor > 0:
        if valor > conta["saldo"]:
            diferenca = valor - conta["saldo"]
            conta["saldo"] = 0
            conta["emprestimo"] -= diferenca
            conta["extrato"] += f"Saque: R$ {valor:.2f} (R$ {diferenca:.2f} emprestados)\n"
        else:
            conta["saldo"] -= valor
            conta["extrato"] += f"Saque: R$ {valor:.2f}\n"

        conta["numero_saques"] += 1
        print(f"Saque realizado com sucesso! Saldo atual: R$ {conta['saldo']:.2f}, Empréstimo disponível: R$ {conta['emprestimo']:.2f}")
    else:
        print("Operação falhou! O valor informado é inválido.")


def exibir_extrato(conta, /):
    print("\n================ EXTRATO ================")
    print("Não foram realizadas movimentações." if not conta["extrato"] else conta["extrato"])
    print(f"\nSaldo: R$ {conta['saldo']:.2f}")
    print(f"Empréstimo disponível: R$ {conta['emprestimo']:.2f}")
    print("==========================================")


numero_conta = 2
AGENCIA = "0001"
LIMITE_SAQUES = 3
LIMITE_SAQUE_VALOR = 500

menu_principal = """

[1] Depositar
[2] Sacar
[3] Extrato
[4] Novo Usuário
[5] Nova Conta
[6] Listar Contas
[7] Sair

=> """

while True:
    opcao = input(menu_principal)

    if opcao == "1":
        conta_selecionada = selecionar_conta(contas)
        if conta_selecionada:
            valor = float(input("Informe o valor do depósito: "))
            depositar(conta_selecionada, valor)

    elif opcao == "2":
        conta_selecionada = selecionar_conta(contas)
        if conta_selecionada:
            valor = float(input("Informe o valor do saque: "))
            sacar(conta=conta_selecionada, valor=valor, limite=LIMITE_SAQUE_VALOR, limite_saques=LIMITE_SAQUES)

    elif opcao == "3":
        conta_selecionada = selecionar_conta(contas)
        if conta_selecionada:
            exibir_extrato(conta_selecionada)

    elif opcao == "4":
        criar_usuario(usuarios)

    elif opcao == "5":
        numero_conta = criar_conta_corrente(AGENCIA, numero_conta, usuarios, contas)

    elif opcao == "6":
        listar_contas(contas)

    elif opcao == "7":
        break

    else:
        print("Operação inválida, por favor selecione novamente a operação desejada.")
```
